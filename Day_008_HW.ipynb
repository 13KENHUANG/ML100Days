```python
app_train.describe()
```

```python
#將三種不同類型的特徵分類:int_feature float_feature object_feature
int_feature=[]
float_feature=[]
object_feature=[]
for dtype,feature in zip(app_train.dtypes,app_train.columns):
    if dtype == 'int64':
        int_feature.append(feature)
    elif dtype == 'float64':
        float_feature.append(feature)
    else:
        object_feature.append(feature)
```

```python
app_train[int_feature].describe()
app_train[int_feature].mean()
app_train[int_feature].std()
```

```python
app_train[float_feature].describe()
app_train[float_feature].mean()
app_train[float_feature].std()
```

```python
app_train[object_feature].describe()
app_train[object_feature].mean()
app_train[object_feature].std()
```

```python
print('計算AMT_INCOME_TOTAL 平均數: ',app_train['AMT_INCOME_TOTAL'].mean())
print('計算AMT_INCOME_TOTAL 標準差: ',app_train['AMT_INCOME_TOTAL'].std())
```

```python
#法1
app_train['OWN_CAR_AGE'].plot.hist(alpha=0.6);
```

```python
#法2
plt.hist(app_train['OWN_CAR_AGE'],color='gold')
plt.title('OWN_CAR_AGE')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
plt.hist(app_train['TARGET'],color='green');
```

```python
plt.bar(['0','1'],app_train['TARGET'].value_counts());
```

```python
#當處理不為數值型態的特徵時，可以使用value_counts來計算各唯一值數量
plt.bar(['F','M','XNA'],app_train['CODE_GENDER'].value_counts())
plt.title('CODE_GENDER')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```
